```python
with open("./vilas_state.pkl",'rb') as file:
        all_data2,stack2,link2=pickle.load(file)
len(all_data2)
```

```python
df=pd.DataFrame(all_data2)
```

```python
df2=df[~df['vlaues'].isnull()]
df22=df2.copy()
```

```python
def viw(x):
    return int(bs(str(x),'html5lib').find('strong').text)
df22['viwers']=df22['viwers'].apply(viw)
```

```python
def loc(x):
    try:
        return (bs(str(x),'html5lib').find('strong',class_="c2b small").text.strip())
    except:
        return (x) 
df22['location']=df22['location'].apply(loc)
```

```python
def date(x):
    try:
        x=bs(str(x),'html5lib').find('span',class_="pdingleft10 brlefte5").text.strip()
        return x[x.find('a'):]
    except Exception as e:
        print (e)
        return(x)
df22['data']=df22['data'].apply(date)
```

```python
def title(x):
    try:
        return (bs(str(x),'html5lib').find('h1',class_="brkword lheight28").text.strip())
    except:
        return (x) 
df22['title']=df22['title'].apply(title)
```

```python
blank_record = {'Bedrooms': np.nan, 'Bathrooms': np.nan, 'Area (m²)': np.nan, 'Type': np.nan, 'Payment Option': np.nan, 'Compound': np.nan, 'Delivery Term': np.nan, 'Amenities': np.nan, 'Furnished': np.nan, 'Delivery Date': np.nan,'Down Payment':np.nan}

def valuess(x):
    record=blank_record.copy()
    try:
        soup=bs(str(x),'html5lib')
        x=[ i.find('tr') for i in soup.find_all('table')]
        z={i.find('th').text.strip():i.find('td').text.strip() for i in x}
        record.update(z)
        try:
            record['Amenities'] = '; '.join([x.replace('\n', '') for x in record['Amenities'].split('\t') if x.replace('\n', '') != ''])
            return pd.Series(record)
        except:
            return pd.Series(record)
    except Exception as e:
        print(e)
        print(len(record))
        return blank_record
    
df22[list(blank_record.keys())]=df22['vlaues'].apply(valuess)
```
