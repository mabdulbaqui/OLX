```python
from  selenium import webdriver 
from bs4 import BeautifulSoup as bs
import re 
from translate import Translator
import string
import re
from time import  sleep
from selenium.common.exceptions import TimeoutException
import pickle
from collections import deque
from selenium.webdriver.common.by import By
import pandas as pd
import sys
sys.setrecursionlimit(10**6)
```

```python
def getfunc(link):
    while True:
        try:
            driver.get(link)
            x=driver.page_source
            return x
        except:
            print("time_out")
            sleep(120)
```

```python
def ads_list(page_Source):
    try:
        soup=bs(page_Source,'html5lib')
        return[i['href'][0:23]+"en/"+i['href'][23:] for i in soup.find_all('a',class_="ads__item__ad--title")]

    except  Exception as e:
        print( str(e),"ad_list")
```

```python
def next_page(page_source):
    try:
        x=bs(page_source,'html5lib').find_all('a',class_=re.compile("pageNextPrev"))
        if len(x)==1:
                if x[0].text.strip() =='Next »':
                    return x[0]['href']
        elif len(x)==2:
            return x[1]['href']
        else:
            return False
    except Exception as e :
        print(str(e)+' next')
```

```python
def extract_data(page_source,i):
    
        res={}
        res.update({"link":i})
        AddSoup=bs(page_source,'html5lib')
        date=AddSoup.find('span',class_="pdingleft10 brlefte5")
        try:
            Fdata=date.get_text()
            Fdate=Fdata.split(',')[1]
            res.update({"date":Fdate})
        except:
            res.update({"date":date})
        ############################################
        try:
            title=AddSoup.find('h1',class_=re.compile("brkword lheight28"))
            Ftitle=title.text.strip()
            res.update({"title":Ftitle})
        except:
            res.update({"title":title})
        ##################################################
        try:
            viwers=AddSoup.find_all('div',class_="pdingtop10")
            Fviwers=viwers[1].find('strong').get_text()
            res.update({"viwers":Fviwers})
        except:
            res.update({"viwers":viwers})
        ###################################################
        try:
            location=AddSoup.find('strong',class_="c2b small")
            Flocation=location.get_text().strip()
            res.update({"location":Flocation})
        except:
            res.update({"location":location})
        ###################################################
        try:
            price=AddSoup.find('strong',class_=re.compile("xxxx-large margintop7"))
            if price['class']=="xxxx-large margintop7 block arranged":
                sign=1
            else:
                sign=0
            Fprice=price.get_text()
            res.update({"Price":Fprice})
            res.update({'sign':sign})
        except:
            res.update({"Price":price})
        ###################################################
        try:
            table=AddSoup.find(class_="details fixed marginbott20 margintop5 full")
            FiledsNames={i.find('th').text.replace('\t', '').strip():i.find('td').text.replace('\t', '').strip() for i in table.find_all('tr')}
            res.update(FiledsNames)
        except:
            res.update({"table":table})
        ###################################################
        
        return  res
```

```python
chrome_options = webdriver.ChromeOptions()
prefs = {"profile.managed_default_content_settings.images": 2}
chrome_options.add_experimental_option("prefs", prefs)
driver = webdriver.Chrome('D:\coding-area\olx_scrapping\chromedriver.exe',chrome_options=chrome_options)

link="https://www.olx.com.eg/en/properties/apartments-duplex-for-rent/"

try:
    with open("state.pkl",'rb') as file:
        all_data, stack, last_link = pickle.load(file)
except:
    all_data, stack, last_link = [], deque(), link

stack.appendleft(last_link)

while stack:
    link=stack.pop()
    last_link=link
    src=getfunc(link)
    print(len(all_data))
    if '/ad/' in link and link not in {x['link'] for x in all_data}:
        data=extract_data(src,link)
        all_data.append(data)
        with open("state.pkl",'wb') as file:
            state = all_data, stack, last_link
            pickle.dump(state, file)
        sleep(3)
    else:
        link=next_page(src)
        if link:
            stack.appendleft(link)
        stack.extendleft(ads_list(src))
```

```python
import numpy as np
series = df['Amenities'].apply(lambda x: [y.strip() for y in x.split('\n') if y.strip() != ''] if type(x) == str else [np.nan])
series
```

```python
amenities = {y for x in series for y in x}
amenities.discard(np.nan)
amenities = list(amenities)
amenities
```

```python
df[amenities] = df['Amenities'].apply(lambda x: pd.Series([y in x for y in amenities] if type(x) == str else [False] * len(amenities)))
df.head()
```
